在绘图中，通常要用到 matplotlib 的 pyplot 模块。这里我总结了一些常用的使用方法。

首先以绘制函数曲线为例

```
import numpy as np 
import matplotlib.pyplot as plt
```

```
x = np.arange(1,11) 
y = 6*x+6 
plt.title("Name: Yusong Luo      Class: 2018211129")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y) 
plt.show()
```

除了绘制连续的函数图像，我们还可以通过“描点”来绘制离散图。下面列举了一些常用用法：

image.png

下面用黑色圆点代替线来绘制上面的函数图像：

```
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,11)
y = 6*x+6
plt.title("Name: Yusong Luo      Class: 2018211129")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y,'ob')
plt.show()
```

除了将线条改为点，我们还可以改变线条的颜色、宽度等：

```
plt.subplot(2,1,1) 
plt.plot(x,y,color='blue',linewidth='5')
plt.subplot(2,1,2)
plt.plot(x,y,color='red',linewidth='2')
```

上面我们一直在使用 pyplot 模块中的 plot() 函数，下面来试试 scatter() 函数。

scatter() 函数常用于绘制散点图：

```
plt.scatter(x,y)
```

同样地，我们可以改变点的颜色和形状：

```
plt.scatter(x,y,color='red',marker='v')
```

另外，我们还可以将两张图绘制在同一个坐标系下，如：

```
plt.plot(x,y)
x2 = np.arange(1,11)
y2 = 5*x2+5
plt.plot(x2,y2)
plt.show()
```

某些情况下，我们还需要限定 x 轴和 y 轴的范围，这时便可以使用 xlim() 函数和 ylim() 函数：

```
# 未限定前
plt.subplot(2,1,1)
plt.plot(x,y)
# 限定后
plt.subplot(2,1,2)
plt.xlim(0,5)
plt.ylim(10,35)
plt.plot(x,y)
```
